```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# ============================================================
# Configuração da região Ω = [a,b] x [c,d]
# ============================================================
a, b = -10, 10
c, d = -8, 8

# ============================================================
# Funções auxiliares
# ============================================================
def segment_curve(x, y):
    """
    Recebe arrays x e y e devolve uma lista de segmentos,
    onde cada segmento é dado por ((x_i, y_i), (x_{i+1}, y_{i+1})).
    """
    segments = []
    for i in range(len(x) - 1):
        p1 = (x[i], y[i])
        p2 = (x[i + 1], y[i + 1])
        segments.append((p1, p2))
    return segments

def plot_segments(ax, segments, color="blue", alpha=0.8, lw=2, label=None):
    """
    Plota os segmentos em um eixo matplotlib.
    """
    first = True
    for (x1, y1), (x2, y2) in segments:
        ax.plot([x1, x2], [y1, y2],
                color=color, alpha=alpha, linewidth=lw,
                label=label if first else None)
        first = False

# ============================================================
# Curva fechada: elipse deslocada
# ============================================================
t_closed = np.linspace(0, 2 * np.pi, 40)
x_closed = 2.5 * np.cos(t_closed) - 2.0
y_closed = 1.5 * np.sin(t_closed) + 2.0

# Fechar explicitamente a curva ligando o último ao primeiro
x_closed = np.append(x_closed, x_closed[0])
y_closed = np.append(y_closed, y_closed[0])

closed_segments = segment_curve(x_closed, y_closed)

# ============================================================
# Curva aberta: senoide paramétrica deslocada
# ============================================================
t_open = np.linspace(-4, 4, 35)
x_open = t_open + 2.0
y_open = 1.2 * np.sin(1.3 * t_open) - 2.5

open_segments = segment_curve(x_open, y_open)

# ============================================================
# União dos pontos segmentados
# ============================================================
union_points = set(zip(x_closed, y_closed)) | set(zip(x_open, y_open))

# ============================================================
# Pontos aleatórios dentro da região, com seed fixa
# ============================================================
rng = np.random.default_rng(42)
n_points = 50

random_x = rng.uniform(a, b, n_points)
random_y = rng.uniform(c, d, n_points)

mx = (a + b) / 2
my = (c + d) / 2
R = 1.0
lam = (d - c) / (b - a)

def distance_point_to_box(px, py, xmin, xmax, ymin, ymax):
    dx = max(xmin - px, 0, px - xmax)
    dy = max(ymin - py, 0, py - ymax)
    return np.hypot(dx, dy)


def point_in_triangle(px, py, tri, eps=1e-12):
    """
    Teste de pertencimento ao triângulo via orientação.
    Inclui a borda.
    """
    A, B, C = tri

    def cross(P, Q, R):
        return (Q[0] - P[0]) * (R[1] - P[1]) - (Q[1] - P[1]) * (R[0] - P[0])

    P = np.array([px, py], dtype=float)

    c1 = cross(A, B, P)
    c2 = cross(B, C, P)
    c3 = cross(C, A, P)

    has_neg = (c1 < -eps) or (c2 < -eps) or (c3 < -eps)
    has_pos = (c1 > eps) or (c2 > eps) or (c3 > eps)

    return not (has_neg and has_pos)

# ============================================================
# Regiões W_i como triângulos
# ============================================================
regions = {
    "W1": np.array([[a, my], [mx, d], [a, d]], dtype=float),
    "W2": np.array([[a, my], [mx, d], [mx, my]], dtype=float),

    "W3": np.array([[mx, d], [b, my], [b, d]], dtype=float),
    "W4": np.array([[mx, d], [b, my], [mx, my]], dtype=float),

    "W5": np.array([[mx, my], [b, my], [mx, c]], dtype=float),
    "W6": np.array([[mx, c], [b, my], [b, c]], dtype=float),

    "W7": np.array([[a, c], [a, my], [mx, c]], dtype=float),
    "W8": np.array([[a, my], [mx, my], [mx, c]], dtype=float),
}

# ordem preferencial para resolver pontos de fronteira
region_order = ["W1", "W2", "W3", "W4", "W5", "W6", "W7", "W8"]

# bounding boxes
region_boxes = {}
for name, poly in regions.items():
    xmin = np.min(poly[:, 0])
    xmax = np.max(poly[:, 0])
    ymin = np.min(poly[:, 1])
    ymax = np.max(poly[:, 1])
    region_boxes[name] = (xmin, xmax, ymin, ymax)

# ============================================================
# Classificação robusta por pertencimento ao triângulo
# ============================================================
def classify_W(px, py):
    for name in region_order:
        if point_in_triangle(px, py, regions[name]):
            return name
    return None

def point_segment_distance(px, py, ax, ay, bx, by):
    """
    Distância de um ponto P=(px,py) ao segmento AB.
    """
    ap = np.array([px - ax, py - ay], dtype=float)
    ab = np.array([bx - ax, by - ay], dtype=float)

    ab2 = np.dot(ab, ab)
    if ab2 == 0:
        return np.hypot(px - ax, py - ay)

    t = np.dot(ap, ab) / ab2
    t = max(0.0, min(1.0, t))

    closest = np.array([ax, ay]) + t * ab
    return np.hypot(px - closest[0], py - closest[1])


def point_triangle_distance(px, py, tri):
    """
    Distância exata de um ponto ao triângulo.
    Se o ponto estiver dentro, a distância é zero.
    """
    if point_in_triangle(px, py, tri):
        return 0.0

    A, B, C = tri
    d1 = point_segment_distance(px, py, A[0], A[1], B[0], B[1])
    d2 = point_segment_distance(px, py, B[0], B[1], C[0], C[1])
    d3 = point_segment_distance(px, py, C[0], C[1], A[0], A[1])
    return min(d1, d2, d3)

def reachable_regions_for_point_exact(px, py, R, regions):
    """
    Retorna apenas as regiões triangulares W_i cuja distância exata ao ponto é <= R.
    """
    result = []
    for name, tri in regions.items():
        dist = point_triangle_distance(px, py, tri)
        if dist <= R:
            result.append(name)
    return result

# ============================================================
# Informações das amostras
# ============================================================
closed_info = []
for i, (x, y) in enumerate(zip(x_closed, y_closed)):
    w = classify_W(x, y)
    reachable = reachable_regions_for_point_exact(x, y, R, regions)
    closed_info.append((i, x, y, w, reachable))

open_info = []
for i, (x, y) in enumerate(zip(x_open, y_open)):
    w = classify_W(x, y)
    reachable = reachable_regions_for_point_exact(x, y, R, regions)
    open_info.append((i, x, y, w, reachable))

# ============================================================
# Plot
# ============================================================
fig, ax = plt.subplots(figsize=(11, 9))

region = Rectangle((a, c), b - a, d - c, fill=False, edgecolor="black", linewidth=2)
ax.add_patch(region)

# linhas dos quadrantes
ax.axvline(mx, color="gray", linestyle="--", linewidth=1)
ax.axhline(my, color="gray", linestyle="--", linewidth=1)

# retas pontilhadas dos W_i
x_left = np.linspace(a, mx, 100)
x_right = np.linspace(mx, b, 100)
ax.plot(x_left,  my + lam * (x_left - a), linestyle=":", linewidth=2, color="black", label="Retas dos $W_i$")
ax.plot(x_left,  my - lam * (x_left - a), linestyle=":", linewidth=2, color="black")
ax.plot(x_right, my - lam * (x_right - b), linestyle=":", linewidth=2, color="black")
ax.plot(x_right, my + lam * (x_right - b), linestyle=":", linewidth=2, color="black")

# curvas
plot_segments(ax, closed_segments, color="royalblue", lw=2.2, label="Curva fechada segmentada")
plot_segments(ax, open_segments, color="darkorange", lw=2.2, label="Curva aberta segmentada")

# amostras
ax.scatter(x_closed, y_closed, color="navy", s=25, zorder=3, label="Amostras curva fechada")
ax.scatter(x_open, y_open, color="orangered", s=25, zorder=3, label="Amostras curva aberta")

# aleatórios
ax.scatter(random_x, random_y, color="forestgreen", s=40, marker="x", zorder=4, label=f"{n_points} pontos aleatórios")

# rótulos
region_centers = {
    "W1": (-7.5,  6.0),
    "W2": (-2.5,  6.0),
    "W3": ( 7.5,  6.0),
    "W4": ( 2.5,  6.0),
    "W5": ( 2.5, -2.0),
    "W6": ( 7.5, -6.0),
    "W7": (-7.5, -6.0),
    "W8": (-2.5, -2.0),
}
for name, (tx, ty) in region_centers.items():
    ax.text(tx, ty, name, fontsize=11, ha="center", va="center", alpha=0.8)

ax.set_xlim(a - 1, b + 1)
ax.set_ylim(c - 1, d + 1)
ax.set_aspect("equal")
ax.set_title(f"Partição em $W_i$ com classificação geométrica consistente, {n_points} pontos aleatórios (R={R})")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, linestyle="--", alpha=0.4)
ax.legend()
plt.show()

# ============================================================
# Mostrar algumas classificações
# ============================================================
reachable_by_point = {name: [] for name in region_order}
print("=== Curva fechada ===")
for i, x, y, w, reachable in closed_info:
    print(f"p{i:02d} = ({x: .3f}, {y: .3f}) | W = {w} | regiões candidatas: {reachable}")
    reachable_by_point[(x,y)] = reachable

print("\n=== Curva aberta ===")
for i, x, y, w, reachable in open_info:
    print(f"p{i:02d} = ({x: .3f}, {y: .3f}) | W = {w} | regiões candidatas: {reachable}")
    reachable_by_point[(x,y)] = reachable
    
print("\n=== Pontos aleatórios ===")
points_by_region = {name: [] for name in region_order}
for i, x, y in zip(range(n_points), random_x, random_y):
    w = classify_W(x, y)
    points_by_region[w].append((x, y))
    
for p in union_points:
    reachable = reachable_by_point.get(p, [])
    for w in reachable:
        for q in points_by_region[w]:
            if np.hypot(p[0] - q[0], p[1] - q[1]) <= R:
                print(f"Ponto ({p[0]:.3f}, {p[1]:.3f}) classificado em W: {w} é alcançável por ponto aleatório ({q[0]:.3f}, {q[1]:.3f})")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# ============================================================
# Configuração da região Ω = [a,b] x [c,d]
# ============================================================
a, b = -10.0, 10.0
c, d = -8.0, 8.0
mx = (a + b) / 2.0
my = (c + d) / 2.0
R = 1.0
lam = (d - c) / (b - a)

# ============================================================
# Funções auxiliares para curvas e plot
# ============================================================
def segment_curve(x, y):
    return [((x[i], y[i]), (x[i + 1], y[i + 1])) for i in range(len(x) - 1)]

def plot_segments(ax, segments, color="blue", alpha=0.8, lw=2, label=None):
    first = True
    for (x1, y1), (x2, y2) in segments:
        ax.plot(
            [x1, x2],
            [y1, y2],
            color=color,
            alpha=alpha,
            linewidth=lw,
            label=label if first else None,
        )
        first = False

# ============================================================
# Geometria escalar
# ============================================================
def distance_point_to_box(px, py, xmin, xmax, ymin, ymax):
    dx = max(xmin - px, 0.0, px - xmax)
    dy = max(ymin - py, 0.0, py - ymax)
    return math.hypot(dx, dy)

def point_segment_distance(px, py, ax, ay, bx, by):
    abx = bx - ax
    aby = by - ay
    apx = px - ax
    apy = py - ay

    ab2 = abx * abx + aby * aby
    if ab2 == 0.0:
        return math.hypot(px - ax, py - ay)

    t = (apx * abx + apy * aby) / ab2
    if t < 0.0:
        t = 0.0
    elif t > 1.0:
        t = 1.0

    cx = ax + t * abx
    cy = ay + t * aby
    return math.hypot(px - cx, py - cy)

def orient(ax, ay, bx, by, px, py):
    return (bx - ax) * (py - ay) - (by - ay) * (px - ax)

def point_in_triangle(px, py, tri, eps=1e-12):
    ax, ay = tri[0]
    bx, by = tri[1]
    cx, cy = tri[2]

    c1 = orient(ax, ay, bx, by, px, py)
    c2 = orient(bx, by, cx, cy, px, py)
    c3 = orient(cx, cy, ax, ay, px, py)

    has_neg = (c1 < -eps) or (c2 < -eps) or (c3 < -eps)
    has_pos = (c1 > eps) or (c2 > eps) or (c3 > eps)
    return not (has_neg and has_pos)

def point_triangle_distance(px, py, tri):
    if point_in_triangle(px, py, tri):
        return 0.0

    ax, ay = tri[0]
    bx, by = tri[1]
    cx, cy = tri[2]

    d1 = point_segment_distance(px, py, ax, ay, bx, by)
    d2 = point_segment_distance(px, py, bx, by, cx, cy)
    d3 = point_segment_distance(px, py, cx, cy, ax, ay)
    return min(d1, d2, d3)

# ============================================================
# Classificação analítica em W_i
# ============================================================
def classify_W(px, py, eps=1e-12):
    # Quadrante superior esquerdo
    if px <= mx and py >= my:
        y_line = my + lam * (px - a)
        return "W1" if py >= y_line - eps else "W2"

    # Quadrante superior direito
    if px >= mx and py >= my:
        y_line = my - lam * (px - b)
        return "W3" if py >= y_line - eps else "W4"

    # Quadrante inferior direito
    if px >= mx and py <= my:
        y_line = my + lam * (px - b)
        return "W6" if py <= y_line + eps else "W5"

    # Quadrante inferior esquerdo
    y_line = my - lam * (px - a)
    return "W7" if py <= y_line + eps else "W8"

# ============================================================
# Regiões W_i como triângulos
# ============================================================
regions = {
    "W1": np.array([[a, my], [mx, d], [a, d]], dtype=float),
    "W2": np.array([[a, my], [mx, d], [mx, my]], dtype=float),
    "W3": np.array([[mx, d], [b, my], [b, d]], dtype=float),
    "W4": np.array([[mx, d], [b, my], [mx, my]], dtype=float),
    "W5": np.array([[mx, my], [b, my], [mx, c]], dtype=float),
    "W6": np.array([[mx, c], [b, my], [b, c]], dtype=float),
    "W7": np.array([[a, c], [a, my], [mx, c]], dtype=float),
    "W8": np.array([[a, my], [mx, my], [mx, c]], dtype=float),
}
region_order = ["W1", "W2", "W3", "W4", "W5", "W6", "W7", "W8"]

# Bounding boxes pré-computadas
region_boxes = {}
for name, tri in regions.items():
    xmin = float(np.min(tri[:, 0]))
    xmax = float(np.max(tri[:, 0]))
    ymin = float(np.min(tri[:, 1]))
    ymax = float(np.max(tri[:, 1]))
    region_boxes[name] = (xmin, xmax, ymin, ymax)

def reachable_regions_for_point_exact(px, py, R, regions, region_boxes):
    """
    Poda por bounding box + distância exata ao triângulo.
    """
    result = []
    for name in region_order:
        xmin, xmax, ymin, ymax = region_boxes[name]
        if distance_point_to_box(px, py, xmin, xmax, ymin, ymax) > R:
            continue
        if point_triangle_distance(px, py, regions[name]) <= R:
            result.append(name)
    return result

# ============================================================
# Grade espacial uniforme para pontos aleatórios
# ============================================================
CELL_SIZE = R

def cell_index(x, y, cell_size=CELL_SIZE):
    ix = int(math.floor((x - a) / cell_size))
    iy = int(math.floor((y - c) / cell_size))
    return ix, iy

def build_spatial_grid(points, cell_size=CELL_SIZE):
    grid = {}
    for qx, qy in points:
        key = cell_index(qx, qy, cell_size)
        grid.setdefault(key, []).append((qx, qy))
    return grid

def candidate_points_from_grid(px, py, grid, cell_size=CELL_SIZE):
    ix, iy = cell_index(px, py, cell_size)
    result = []
    for dx in (-1, 0, 1):
        for dy in (-1, 0, 1):
            key = (ix + dx, iy + dy)
            if key in grid:
                result.extend(grid[key])
    return result

# ============================================================
# Curvas
# ============================================================
t_closed = np.linspace(0, 2 * np.pi, 40)
x_closed = 2.5 * np.cos(t_closed) - 2.0
y_closed = 1.5 * np.sin(t_closed) + 2.0
x_closed = np.append(x_closed, x_closed[0])
y_closed = np.append(y_closed, y_closed[0])
closed_segments = segment_curve(x_closed, y_closed)

t_open = np.linspace(-4, 4, 35)
x_open = t_open + 2.0
y_open = 1.2 * np.sin(1.3 * t_open) - 2.5
open_segments = segment_curve(x_open, y_open)

# União dos pontos amostrados
union_points = np.vstack([
    np.column_stack([x_closed, y_closed]),
    np.column_stack([x_open, y_open]),
])

# ============================================================
# Pontos aleatórios
# ============================================================
rng = np.random.default_rng(42)
n_points = 50
random_x = rng.uniform(a, b, n_points)
random_y = rng.uniform(c, d, n_points)
random_points = list(zip(random_x, random_y))

# Índices por região e por grade
points_by_region = {name: [] for name in region_order}
for qx, qy in random_points:
    w = classify_W(qx, qy)
    points_by_region[w].append((qx, qy))

grid = build_spatial_grid(random_points, CELL_SIZE)

# ============================================================
# Informações das amostras
# ============================================================
closed_info = []
for i, (x, y) in enumerate(zip(x_closed, y_closed)):
    w = classify_W(x, y)
    reachable = reachable_regions_for_point_exact(x, y, R, regions, region_boxes)
    closed_info.append((i, x, y, w, reachable))

open_info = []
for i, (x, y) in enumerate(zip(x_open, y_open)):
    w = classify_W(x, y)
    reachable = reachable_regions_for_point_exact(x, y, R, regions, region_boxes)
    open_info.append((i, x, y, w, reachable))

# ============================================================
# Busca eficiente de pontos aleatórios que alcançam cada amostra
# ============================================================
reachable_hits = []

for px, py in union_points:
    reachable_regions = set(reachable_regions_for_point_exact(px, py, R, regions, region_boxes))

    # candidatos locais pela grade
    local_candidates = candidate_points_from_grid(px, py, grid, CELL_SIZE)

    for qx, qy in local_candidates:
        # poda adicional por região
        qw = classify_W(qx, qy)
        if qw not in reachable_regions:
            continue

        if math.hypot(px - qx, py - qy) <= R:
            reachable_hits.append(((px, py), qw, (qx, qy)))

# ============================================================
# Plot
# ============================================================
fig, ax = plt.subplots(figsize=(11, 9))

region = Rectangle((a, c), b - a, d - c, fill=False, edgecolor="black", linewidth=2)
ax.add_patch(region)

# linhas dos quadrantes
ax.axvline(mx, color="gray", linestyle="--", linewidth=1)
ax.axhline(my, color="gray", linestyle="--", linewidth=1)

# retas pontilhadas dos W_i
x_left = np.linspace(a, mx, 100)
x_right = np.linspace(mx, b, 100)
ax.plot(x_left,  my + lam * (x_left - a), linestyle=":", linewidth=2, color="black", label="Retas dos $W_i$")
ax.plot(x_left,  my - lam * (x_left - a), linestyle=":", linewidth=2, color="black")
ax.plot(x_right, my - lam * (x_right - b), linestyle=":", linewidth=2, color="black")
ax.plot(x_right, my + lam * (x_right - b), linestyle=":", linewidth=2, color="black")

# curvas
plot_segments(ax, closed_segments, color="royalblue", lw=2.2, label="Curva fechada segmentada")
plot_segments(ax, open_segments, color="darkorange", lw=2.2, label="Curva aberta segmentada")

# amostras
ax.scatter(x_closed, y_closed, color="navy", s=25, zorder=3, label="Amostras curva fechada")
ax.scatter(x_open, y_open, color="orangered", s=25, zorder=3, label="Amostras curva aberta")

# aleatórios
ax.scatter(random_x, random_y, color="forestgreen", s=40, marker="x", zorder=4, label=f"{n_points} pontos aleatórios")

# rótulos
region_centers = {
    "W1": (-7.5,  6.0),
    "W2": (-2.5,  6.0),
    "W3": ( 7.5,  6.0),
    "W4": ( 2.5,  6.0),
    "W5": ( 2.5, -2.0),
    "W6": ( 7.5, -6.0),
    "W7": (-7.5, -6.0),
    "W8": (-2.5, -2.0),
}
for name, (tx, ty) in region_centers.items():
    ax.text(tx, ty, name, fontsize=11, ha="center", va="center", alpha=0.8)

ax.set_xlim(a - 1, b + 1)
ax.set_ylim(c - 1, d + 1)
ax.set_aspect("equal")
ax.set_title(f"Partição em $W_i$ com busca otimizada, {n_points} pontos aleatórios (R={R})")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, linestyle="--", alpha=0.4)
ax.legend()
plt.show()

# ============================================================
# Mostrar resultados
# ============================================================
print("=== Curva fechada ===")
for i, x, y, w, reachable in closed_info:
    print(f"p{i:02d} = ({x: .3f}, {y: .3f}) | W = {w} | regiões candidatas: {reachable}")

print("\n=== Curva aberta ===")
for i, x, y, w, reachable in open_info:
    print(f"p{i:02d} = ({x: .3f}, {y: .3f}) | W = {w} | regiões candidatas: {reachable}")

print("\n=== Alcances encontrados ===")
for (px, py), w, (qx, qy) in reachable_hits:
    print(
        f"Ponto ({px:.3f}, {py:.3f}) pode ser alcançado por "
        f"({qx:.3f}, {qy:.3f}) na região {w}"
    )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# ============================================================
# Configuração da região Ω = [a,b] x [c,d]
# ============================================================
a, b = -10, 10
c, d = -8, 8

mx = (a + b) / 2   # 0.0
my = (c + d) / 2   # 0.0
R  = 1.0
lam = (d - c) / (b - a)   # 0.8  (inclinação das diagonais)

# ============================================================
# Regiões W_i como triângulos  (definidas uma única vez)
# ============================================================
regions = {
    "W1": np.array([[a, my], [mx, d], [a, d]], dtype=float),
    "W2": np.array([[a, my], [mx, d], [mx, my]], dtype=float),
    "W3": np.array([[mx, d], [b, my], [b, d]], dtype=float),
    "W4": np.array([[mx, d], [b, my], [mx, my]], dtype=float),
    "W5": np.array([[mx, my], [b, my], [mx, c]], dtype=float),
    "W6": np.array([[mx, c], [b, my], [b, c]], dtype=float),
    "W7": np.array([[a, c], [a, my], [mx, c]], dtype=float),
    "W8": np.array([[a, my], [mx, my], [mx, c]], dtype=float),
}
region_order = ["W1", "W2", "W3", "W4", "W5", "W6", "W7", "W8"]

# Bounding boxes pré-computadas
region_boxes = {
    name: (
        float(np.min(poly[:, 0])), float(np.max(poly[:, 0])),
        float(np.min(poly[:, 1])), float(np.max(poly[:, 1])),
    )
    for name, poly in regions.items()
}

# ============================================================
# OTIMIZAÇÃO 1 — classify_W em O(1) usando geometria analítica
#
# As 4 diagonais dividem o retângulo em 8 triângulos.
# Em vez de testar pertencimento a cada triângulo, basta olhar:
#   • qual quadrante o ponto está  (sinal de dx, dy)
#   • se está mais perto do centro ou do canto  (sinal de r)
# ============================================================
def classify_W(px, py):
    dx = px - mx
    dy = py - my
    # distância "diagonal": positivo → lado do canto; negativo → lado do centro
    r  = abs(dy) - lam * abs(dx)

    if dx <= 0 and dy >= 0:          # quadrante superior-esquerdo
        return "W1" if r >= 0 else "W2"
    elif dx >= 0 and dy >= 0:        # quadrante superior-direito
        return "W3" if r >= 0 else "W4"
    elif dx >= 0 and dy <= 0:        # quadrante inferior-direito
        return "W6" if r <= 0 else "W5"
    else:                            # quadrante inferior-esquerdo
        return "W7" if r <= 0 else "W8"


# ============================================================
# OTIMIZAÇÃO 2 — point_in_triangle sem alocação NumPy
#
# Usa aritmética pura com floats escalares.
# Elimina np.array([px, py]) a cada chamada.
# ============================================================
def point_in_triangle(px, py, tri, eps=1e-12):
    """Teste de orientação (cross product). Inclui bordas."""
    (ax, ay), (bx, by), (cx, cy) = tri[0], tri[1], tri[2]

    c1 = (bx - ax) * (py - ay) - (by - ay) * (px - ax)
    c2 = (cx - bx) * (py - by) - (cy - by) * (px - bx)
    c3 = (ax - cx) * (py - cy) - (ay - cy) * (px - cx)

    has_neg = (c1 < -eps) or (c2 < -eps) or (c3 < -eps)
    has_pos = (c1 >  eps) or (c2 >  eps) or (c3 >  eps)
    return not (has_neg and has_pos)


# ============================================================
# OTIMIZAÇÃO 3 — distâncias sem alocações extras
# ============================================================
def _point_segment_dist_sq(px, py, ax, ay, bx, by):
    """Distância ao quadrado de (px,py) ao segmento AB (sem sqrt)."""
    abx, aby = bx - ax, by - ay
    ab2 = abx * abx + aby * aby
    if ab2 == 0.0:
        return (px - ax)**2 + (py - ay)**2
    t = max(0.0, min(1.0, ((px - ax) * abx + (py - ay) * aby) / ab2))
    ex, ey = ax + t * abx - px, ay + t * aby - py
    return ex * ex + ey * ey


def point_triangle_distance(px, py, tri):
    """
    Distância exata de um ponto ao triângulo.
    Retorna 0.0 se o ponto estiver dentro.
    """
    if point_in_triangle(px, py, tri):
        return 0.0
    (ax, ay), (bx, by), (cx, cy) = tri[0], tri[1], tri[2]
    d1 = _point_segment_dist_sq(px, py, ax, ay, bx, by)
    d2 = _point_segment_dist_sq(px, py, bx, by, cx, cy)
    d3 = _point_segment_dist_sq(px, py, cx, cy, ax, ay)
    return np.sqrt(min(d1, d2, d3))


# ============================================================
# OTIMIZAÇÃO 4 — reachable_regions com rejeição por bbox
# ============================================================
def reachable_regions_for_point(px, py, R, regions):
    """
    Regiões W_i cuja distância exata ao ponto é <= R.
    Rejeição rápida por bounding box antes de calcular distância exata.
    """
    result = []
    R2 = R * R
    for name, tri in regions.items():
        xmin, xmax, ymin, ymax = region_boxes[name]
        if px < xmin - R or px > xmax + R or py < ymin - R or py > ymax + R:
            continue
        if point_triangle_distance(px, py, tri) <= R:
            result.append(name)
    return result


# ============================================================
# Funções auxiliares de curva (inalteradas na lógica,
# mas segment_curve usa NumPy em vez de loop Python)
# ============================================================
def segment_curve(x, y):
    """Devolve array (N-1, 2, 2) com todos os segmentos."""
    p1 = np.stack([x[:-1], y[:-1]], axis=1)
    p2 = np.stack([x[1:],  y[1:]], axis=1)
    return list(zip(map(tuple, p1), map(tuple, p2)))

def plot_segments(ax, segments, color="blue", alpha=0.8, lw=2, label=None):
    first = True
    for (x1, y1), (x2, y2) in segments:
        ax.plot([x1, x2], [y1, y2],
                color=color, alpha=alpha, linewidth=lw,
                label=label if first else None)
        first = False


# ============================================================
# Curvas
# ============================================================
t_closed = np.linspace(0, 2 * np.pi, 40)
x_closed = np.append(2.5 * np.cos(t_closed) - 2.0, 2.5 * np.cos(t_closed[0]) - 2.0)
y_closed = np.append(1.5 * np.sin(t_closed) + 2.0, 1.5 * np.sin(t_closed[0]) + 2.0)
closed_segments = segment_curve(x_closed, y_closed)

t_open = np.linspace(-4, 4, 35)
x_open = t_open + 2.0
y_open = 1.2 * np.sin(1.3 * t_open) - 2.5
open_segments = segment_curve(x_open, y_open)

# ============================================================
# Pontos aleatórios
# ============================================================
rng = np.random.default_rng(42)
n_points = 50
random_x = rng.uniform(a, b, n_points)
random_y = rng.uniform(c, d, n_points)

# ============================================================
# OTIMIZAÇÃO 5 — classificação e reachable unificadas
#
# Uma única passagem sobre cada conjunto de pontos.
# Evita dois loops separados idênticos.
# ============================================================
def compute_point_info(xs, ys):
    """Retorna lista de (i, x, y, region, reachable) para cada ponto."""
    info = []
    for i, (x, y) in enumerate(zip(xs, ys)):
        w        = classify_W(x, y)
        reachable = reachable_regions_for_point(x, y, R, regions)
        info.append((i, x, y, w, reachable))
    return info

closed_info = compute_point_info(x_closed, y_closed)
open_info   = compute_point_info(x_open,   y_open)

# ============================================================
# OTIMIZAÇÃO 6 — lookup de reachable por índice (evita dict de floats)
#
# Substitui  reachable_by_point = {(x,y): reachable}
# por listas indexadas, eliminando comparações de float.
# ============================================================
all_pts_x = np.concatenate([x_closed, x_open])
all_pts_y = np.concatenate([y_closed, y_open])
all_reachable = [info[4] for info in closed_info] + [info[4] for info in open_info]

# ============================================================
# Classificação dos pontos aleatórios agrupados por região
# ============================================================
points_by_region = {name: [] for name in region_order}
for x, y in zip(random_x, random_y):
    w = classify_W(x, y)
    points_by_region[w].append((x, y))

# Pré-converte para arrays NumPy para operações vetorizadas
points_by_region_arr = {
    name: np.array(pts) if pts else np.empty((0, 2))
    for name, pts in points_by_region.items()
}

# ============================================================
# Plot
# ============================================================
fig, ax = plt.subplots(figsize=(11, 9))

rect = Rectangle((a, c), b - a, d - c, fill=False, edgecolor="black", linewidth=2)
ax.add_patch(rect)

ax.axvline(mx, color="gray", linestyle="--", linewidth=1)
ax.axhline(my, color="gray", linestyle="--", linewidth=1)

x_left  = np.linspace(a, mx, 100)
x_right = np.linspace(mx, b, 100)
kw_diag = dict(linestyle=":", linewidth=2, color="black")
ax.plot(x_left,  my + lam * (x_left  - a), label="Retas dos $W_i$", **kw_diag)
ax.plot(x_left,  my - lam * (x_left  - a), **kw_diag)
ax.plot(x_right, my - lam * (x_right - b), **kw_diag)
ax.plot(x_right, my + lam * (x_right - b), **kw_diag)

plot_segments(ax, closed_segments, color="royalblue",  lw=2.2, label="Curva fechada segmentada")
plot_segments(ax, open_segments,   color="darkorange", lw=2.2, label="Curva aberta segmentada")

ax.scatter(x_closed,  y_closed,  color="navy",        s=25, zorder=3, label="Amostras curva fechada")
ax.scatter(x_open,    y_open,    color="orangered",    s=25, zorder=3, label="Amostras curva aberta")
ax.scatter(random_x,  random_y,  color="forestgreen",  s=40, zorder=4, marker="x", label=f"{n_points} pontos aleatórios")

region_centers = {
    "W1": (-7.5,  6.0), "W2": (-2.5,  6.0),
    "W3": ( 7.5,  6.0), "W4": ( 2.5,  6.0),
    "W5": ( 2.5, -2.0), "W6": ( 7.5, -6.0),
    "W7": (-7.5, -6.0), "W8": (-2.5, -2.0),
}
for name, (tx, ty) in region_centers.items():
    ax.text(tx, ty, name, fontsize=11, ha="center", va="center", alpha=0.8)

ax.set_xlim(a - 1, b + 1)
ax.set_ylim(c - 1, d + 1)
ax.set_aspect("equal")
ax.set_title(f"Partição em $W_i$ — classificação geométrica otimizada, {n_points} pontos aleatórios (R={R})")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, linestyle="--", alpha=0.4)
ax.legend()
plt.show()

# ============================================================
# Impressão das classificações
# ============================================================
print("=== Curva fechada ===")
for i, x, y, w, reachable in closed_info:
    print(f"p{i:02d} = ({x: .3f}, {y: .3f}) | W = {w} | candidatas: {reachable}")

print("\n=== Curva aberta ===")
for i, x, y, w, reachable in open_info:
    print(f"p{i:02d} = ({x: .3f}, {y: .3f}) | W = {w} | candidatas: {reachable}")

# ============================================================
# OTIMIZAÇÃO 7 — loop final totalmente vetorizado
#
# Original:  O(|union| × |regiões| × |aleatórios|)  com loops Python
# Novo:      operação NumPy por região, sem loop sobre aleatórios
# ============================================================
print("\n=== Alcançabilidade: pontos de curva → aleatórios ===")
for idx, (px, py, reachable) in enumerate(zip(all_pts_x, all_pts_y, all_reachable)):
    for w in reachable:
        arr = points_by_region_arr[w]
        if arr.size == 0:
            continue
        # distância vetorizada a todos os aleatórios da região de uma vez
        dists = np.hypot(arr[:, 0] - px, arr[:, 1] - py)
        mask  = dists <= R
        for q in arr[mask]:
            print(
                f"Ponto ({px:.3f}, {py:.3f}) em {w} "
                f"alcança aleatório ({q[0]:.3f}, {q[1]:.3f})"
            )
```
