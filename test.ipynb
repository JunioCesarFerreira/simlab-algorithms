```python
import random

def random_points_in_rectangle(
    p1: tuple[float, float],
    p2: tuple[float, float],
    N: int
) -> list[tuple[float, float]]:
    if N <= 0:
        return []

    x_min = min(p1[0], p2[0])
    x_max = max(p1[0], p2[0])
    y_min = min(p1[1], p2[1])
    y_max = max(p1[1], p2[1])

    points = [
        (
            random.uniform(x_min, x_max),
            random.uniform(y_min, y_max)
        )
        for _ in range(N)
    ]

    return points
```

```python
import matplotlib.pyplot as plt

points = [(0,0)] + random_points_in_rectangle((-100, -100), (100, 100), 20)

positions: dict[int, tuple[float, float]] = {}

for i, p in enumerate(points):
    positions[i] = p
    print(f"Node {i}: {p}")
    
plt.figure(figsize=(10,10))
for node,(x,y) in positions.items():
    plt.scatter(x,y,color="lightblue",s=200,zorder=2,edgecolor="black",linewidth=1)
    plt.text(x, y,str(node), ha='center', va='center')
plt.title("points")
plt.gca().set_aspect('equal')
plt.show()

R = 50
edges = []

for i in range(len(points)):
    for j in range(i+1, len(points)):
        p1 = points[i]
        p2 = points[j]
        distance = ((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2) ** 0.5
        if distance <= R:
            edges.append((i, j))
            
plt.figure(figsize=(10,10))
for node,(x,y) in positions.items():
    plt.plot(x,y,'o', markersize=12,color="lightblue")
    plt.text(x,y, str(node), ha='center', va='center', color='black') 
for i, j in edges:
    x1, y1 = positions[i]
    x2, y2 = positions[j]
    plt.plot([x1, x2], [y1, y2], color="gray", zorder=1)
    
plt.title("edges")
plt.gca().set_aspect('equal')
plt.show()
```
