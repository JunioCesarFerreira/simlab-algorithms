```python
import random

# Constante: número de seeds a serem geradas
NUMBER_OF_SEEDS = 10  # Altere este valor conforme necessário

# Gera seeds aleatórias no intervalo de 0 a 2^32 - 1
MAX_32BIT_UNSIGNED = 2**32 - 1  # 4294967295

seeds = [random.randint(0, MAX_32BIT_UNSIGNED) for _ in range(NUMBER_OF_SEEDS)]

# Exibe os resultados
print(f"Geradas {NUMBER_OF_SEEDS} seeds no intervalo [0, {MAX_32BIT_UNSIGNED}]:\n")
for i, seed in enumerate(seeds, 1):
    print(f"Seed {i:2d}: {seed:11d} (0x{seed:08X})")
```
